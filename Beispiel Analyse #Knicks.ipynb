```python
# ANALYSIS #KNICKS
import json
import pandas as pd

# Pfad zur zusammengefÃ¼hrten JSON-Datei
#merged_file_path = r'C:\Users\yanni\Desktop\project\#Knicks\merged_file.json'
merged_file_path = r'C:\Users\yanni\Desktop\project\#Knicks_DONE\merged_file.json'


# Laden der JSON-Daten
with open(merged_file_path, 'r') as f:
    data = json.load(f)

# Nur relevante Spalten behalten und umbenennen
relevant_columns = ['date', 'rawContent', 'hashtags', 'replyCoun', 'likeCount', 'retweetCount','viewCount']
data = [{col: row.get(col) for col in relevant_columns} for row in data]
data = pd.DataFrame(data)
#
data = data.rename(columns={'date': 'Date', 'rawContent': 'Content', 'replyCoun': 'Replies', 'likeCount': 'Likes', 'retweetCount': 'Retweets', 'viewCount': 'Views'})

# SÃ¤ubern des Inhalts
def clean_content(text):
    # Beispiel: Entfernen von [redacted]
    text = text.replace('[redacted]', '')
    return text

# Anwendung von clean_content Funktion auf Tweet Texte
data['Content'] = data['Content'].apply(clean_content)

# Konvertierung von datetime zu date (Nur noch Tage, keine Uhrzeiten mehr)
data['Date'] = pd.to_datetime(data['Date']).dt.date

# Tweet Texte alles in Kleinbuchstaben
for i in range(len(data)):
    data.loc[i, "Content"] = data.loc[i, "Content"].lower()
data



data
```

```python
# Anzahl Tweets pro Tag
pd.set_option('display.max_rows', None)
tweet_count = data.groupby('Date').agg(tweet_count=('Content', 'count'))
tweet_count
```

```python
import re
# Regex expression fÃ¼r Keyword (nach was soll er suchen)
regex = r"\b(derrick|rose)\b"

# Schleife durch jede Zeile
for i in range(len(data)):
    
    #  Anzahl Treffer in Tweet (hier sucht er)
    data.loc[i, "count_kw"] = len(re.findall(regex, data.loc[i, "Content"]))

# Anzahl Treffer von Keyword an einem Tag

pd.options.display.max_rows = 400  # Zeige bis zu 100 Zeilen an

kw_count = data.groupby('Date').sum("count_kw")
kw_count
```
